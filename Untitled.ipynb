```python
import tensorflow as tf
import cv2 
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_arr = cv2.imread("train/angry/training_3908.jpg")
```

```python
plt.imshow(img_arr)
```

```python
datadirectory = "train/"
```

```python
Classes = ["angry","disgust","fear","happy","neutral","sad","surprise"]
```

```python
for category in Classes:
    path = os.path.join(datadirectory, category)
    for img in os.listdir(path):
        img_arr = cv2.imread(os.path.join(path,img))
        plt.imshow(cv2.cvtColor(img_arr, cv2.COLOR_BGR2RGB))
        plt.show
        break
    break
    
```

```python
imgsize= 224
new_arr = cv2.resize(img_arr, (imgsize,imgsize))
plt.imshow(cv2.cvtColor(new_arr, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
training_data = []

def clean_data():
    for category in Classes:
        path = os.path.join(datadirectory, category)
        class_num = Classes.index(category)
        for img in os.listdir(path):
            try:
                img_arr = cv2.imread(os.path.join(path,img))
                new_arr = cv2.resize(img_arr, (imgsize,imgsize))
                training_data.append([new_arr, class_num])
            except Exception as e:
                    pass    
```

```python
clean_data()
```

```python
print(len(training_data))
```

```python
import random
```

```python
random.shuffle(training_data)
```

```python
X = []
y = []

for features,labels in training_data:
    X.append(features)
    y.append(labels)
    
X = np.array(X).reshape(-1 , imgsize , imgsize , 3)
```

```python
X.shape
```

```python
Y = np.array(y)
```

```python
X = X/255
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = tf.keras.applications.MobileNetV2()
```

```python
model.summary()
```

```python
base_input = model.layers[0].input
```

```python
base_output = model.layers[-2].output
```

```python
final_output = layers.Dense(128)(base_output)
final_output = layers.Activation('relu')(final_output)
final_output = layers.Dense(64)(final_output)
final_output = layers.Activation('relu')(final_output)
final_output = layers.Dense(7, activation='softmax')(final_output)
```

```python
final_output
```

```python
new_model = keras.Model(inputs = base_input, outputs = final_output)
```

```python
new_model.summary()
```

```python
new_model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
new_model.fit(X,Y, epochs=25)
```

```python
new_model.save('my_model_v3')
```

```python
new_model = tf.keras.models.load_model('my_model_v3')
```
